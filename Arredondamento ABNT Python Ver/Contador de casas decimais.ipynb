```python
def contar_casas_decimais(numero):
    """
    Conta o número de casas decimais de um número decimal.
    
    :param numero: O número cujo número de casas decimais será contado.
    :return: Número de casas decimais.
    """
    numero_str = str(numero).split('.')  # Divide o número pela parte inteira e decimal
    if len(numero_str) > 1:  # Verifica se existe uma parte decimal
        return len(numero_str[1])  # Retorna o comprimento da parte decimal
    return 0  # Se não houver parte decimal, retorna 0

# Exemplo de uso:
numero = 123.456700088
resultado = contar_casas_decimais(numero)
print(f"O número de casas decimais é: {resultado}")  # Saída: 5
```
